```python
import pandas as pd
import numpy as np
import seaborn as sns
import datetime
```

```python
from sodapy import Socrata

# Unauthenticated client only works with public data sets. Note 'None'
# in place of application token, and no username or password:
client = Socrata("data.cityofnewyork.us", None)

# Example authenticated client (needed for non-public datasets):
# client = Socrata(data.cityofnewyork.us,
#                  MyAppToken,
#                  userame="user@example.com",
#                  password="AFakePassword")

num_loops = 40
loop_size = 50000
data = []
# First 2000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
for i in range(num_loops):
    results = client.get("uacg-pexx",
                     select="dropoff_latitude, dropoff_longitude, fare_amount, pickup_latitude, pickup_longitude, tpep_dropoff_datetime, tpep_pickup_datetime, trip_distance",
                     where="(dropoff_latitude>40.64 and dropoff_latitude<40.65) and (dropoff_longitude>-73.795 and dropoff_longitude<-73.775)", 
                     order="tpep_pickup_datetime ASC",
                     limit=loop_size,
                     offset=loop_size * i)
    print("\r> Loop number: {}".format(i))
    for result in results:
        data.append(result)
        
# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(data)
```

```python
print(results_df.shape)
results_df = results_df[results_df["trip_distance"] != "0"]
print(results_df.shape)
```

```python
np.mean(results_df.trip_distance.astype("float"))
```

```python
results_df.tpep_dropoff_datetime = pd.to_datetime(results_df.tpep_dropoff_datetime)
results_df.tpep_pickup_datetime = pd.to_datetime(results_df.tpep_pickup_datetime)
```

```python
results_df["travelTime"] = pd.to_timedelta(results_df.tpep_dropoff_datetime - results_df.tpep_pickup_datetime)
results_df["travelTime"]
```

```python
results_df.to_csv("../output/JFK_data.csv")
```
